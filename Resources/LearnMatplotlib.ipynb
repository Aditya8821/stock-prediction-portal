```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
x=['Jan','Feb','Mar','Apr']
y=[200,100,500,400]
```

# Types of Plot

## Line Plot

```python
plt.plot(x,y);
plt.title('Monthly revenue');
plt.xlabel('Month');
plt.ylabel('Revenue');
```

## Bar Plot

```python
plt.figure(figsize=(10,6))     # It's Optional i.e for defining the size of bars.
plt.bar(x,y);
plt.title('Monthly revenue');
plt.xlabel('Month');
plt.ylabel('Revenue');
```

## Histogram

```python
Marks=np.random.randint(0,100,50)   # Create random marks for 50 students within 0-100 range
plt.hist(Marks);
```

A histogram shows the frequency (or count) of data values within certain ranges (bins).
The bins parameter lets you decide how many intervals (or their exact edges) you want.

```python
plt.hist(Marks, bins=2); # So this below hist shows that 20 students have marks ranging btw (5, 55) and 30 students have marks btw (55, 98).
```

## Scatter Plot

```python
plt.scatter(x,y);
```

## Legends

```python
days=[1,2,3,4,5]
original_price=[10,20,30,40,50]
predicted_price=[12,18,30,42,48]
```

```python
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.plot(days, predicted_price, 'r-o' , label='Predicted Price')
plt.legend()
plt.title("Plot using Legends")
plt.xlabel("Day")
plt.ylabel("Stock Price")
```

## Subplot

```python
plt.subplot(3,1,1)
plt.plot(days, original_price, 'b-o', label='Original Price')
plt.legend()
plt.title("Plot using Legends")
plt.xlabel("Day")
plt.ylabel("Stock Price")


plt.subplot(3,1,2)
plt.plot(days, predicted_price, 'r-o' , label='Predicted Price')
plt.legend()
plt.title("Plot using Legends")
plt.xlabel("Day")
plt.ylabel("Stock Price")

price_diff = [o-p for o,p in zip(original_price, predicted_price)]
plt.subplot(3,1,3)
plt.plot(days, price_diff, color='purple', label='Price Difference')
plt.legend()
plt.title("Price Difference")
plt.xlabel("Day")
plt.ylabel("Price Diff")


plt.tight_layout()
```

## Saving the plot image

```python
plt.subplot(3,1,1)
plt.plot(days, original_price, 'b-o', label='Original Price');
plt.legend();
plt.title("Plot using Legends");
plt.xlabel("Day");
plt.ylabel("Stock Price");
plt.savefig("Example Plot image save", dpi=300)
```
